TP2- Pressure on the compressor (bar).<br>
TP3- Pressure generated at the pneumatic panel (bar).<br>
H1- Valve that is activated when the pressure read by the pressure switch of the command is above the operating pressure of 10.2bar (bar).<br>
DV_pressure- Pressure exerted due to pressure drop generated when air dryers towers discharge the water. When it is equal to zero, the compressor is working under load (bar).<br>
Reservoirs- Pressure inside the air tanks installed on the trains (bar).<br>
Oil_Temperature- Temperature of the oil present on the compressor (°C).<br>
Flowmeter- Airfow was measured on the pneumatic control panel (m3/h).<br>
Motor_Current- Motor’s current, which should present the following values: (i) close to 0A when the compressor turns of; (ii) close to 4A when the compressor is working ofoaded; and (iii) close to 7A when the compressor is operating under load (A);

```python
# convert the time column to a datetime object
df['timestamp'] = pd.to_datetime(df['timestamp'])

# extract the month, day, hour, minute, and second into separate columns
df['Month'] = df['timestamp'].dt.strftime('%m')
df['Day'] = df['timestamp'].dt.strftime('%d')
df['Hour'] = df['timestamp'].dt.strftime('%H')
df['Minute'] = df['timestamp'].dt.strftime('%M')
df['Second'] = df['timestamp'].dt.strftime('%S')
```

```python
df.dtypes
```

```python
df['Month'] = df['Month'].astype(float)
df['Day'] = df['Day'].astype(float)
df['Hour'] = df['Hour'].astype(float)
df['Minute'] = df['Minute'].astype(float)
df['Second'] = df['Second'].astype(float)
```

```python
df = df.drop('timestamp', axis=1)
```
